```python
import numpy as np
import matplotlib.pyplot as plt

# Styling untuk plot
plt.style.use('seaborn-v0_8-poster')
# %matplotlib inline  # Hanya untuk Jupyter Notebook, bisa dihapus kalau tidak pakai notebook

# Define parameters
f = lambda t, s: np.exp(-t)   # ODE: ds/dt = e^(-t)
h = 0.1                       # Step size
t = np.arange(0, 1 + h, h)    # Grid dari 0 ke 1
s0 = 1                        # Initial condition

# Euler method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])  # Euler's method

# Exact solution: s(t) = 1 - e^(-t)
s_exact = 1 - np.exp(-t)

# Plot
plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate (Euler)')
plt.plot(t, s_exact, 'g-', label='Exact Solution')
plt.title('Approximate vs Exact Solution for ds/dt = e^(-t)')
plt.xlabel('t')
plt.ylabel('s(t)')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.01  # step size
t = np.arange(0, 1 + h, h)  # Numerical grid
s0 = 1  # initial condition

# Explicit Euler method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g-', label='Exact')
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')
f = lambda x, y: x**2 - 4*y

h = 0.01                      # step size
x = np.arange(0, 1 + h, h)    # grid dari 0 ke 1
y0 = 1                        # initial condition y(0) = 1

# Inisialisasi array y
y = np.zeros(len(x))
y[0] = y0

# Metode Euler eksplisit
for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])

# Output nilai pendekatan y(0.1)
index_01 = int(0.1 / h)  # indeks untuk x = 0.1
print(f"Pendekatan y(0.1) ≈ {y[index_01]:.6f}")

# Plot hasil
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo--', label='Pendekatan (Euler)')
plt.title("Solusi Pendekatan y' = x^2 - 4y dengan Euler Method")
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')
f = lambda t, y: -4 * y

h = 0.1
t = np.arange(0, 1 + h, h)   # grid
y0 = 1                       # y(0) = 1

# Inisialisasi array
y = np.zeros(len(t))
y[0] = y0

# Euler Method
for i in range(len(t) - 1):
    y[i + 1] = y[i] + h * f(t[i], y[i])

y_exact = np.exp(-4 * t)

# Plot hasil
plt.figure(figsize=(12, 8))
plt.plot(t, y, 'bo--', label='Approximate')
plt.plot(t, y_exact, 'g-', label='Exact')
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
```
