```python
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x,y):
  n = len(y)
  coef = np.zeros([n, n])
  coef[:, 0] = y

  for j in range(1,n):
    for i in range(n - j):
      coef[i][j] = (coef[i + 1][j - 1] - coef[i][j - 1]) / (x[i + j] - x[i])
  return coef

def newton_poly(coef, x_data, x):
  n = len(x_data) - 1
  p = coef[n]

  for k in range(1, n + 1):
    p = coef[n - k] + (x - x_data[n - k]) * p

    return p
```

```python
import numpy as np
import matplotlib.pyplot as plt

def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])  # Membuat tabel kosong untuk divided differences
    coef[:, 0] = y  # Kolom pertama berisi nilai y

    for j in range(1, n):  # Loop untuk mengisi kolom-kolom berikutnya
        for i in range(n - j):
            coef[i][j] = (coef[i + 1][j - 1] - coef[i][j - 1]) / (x[i + j] - x[i])

    return coef
def newton_poly(coef, x_data, x):
    n = len(x_data) - 1
    p = coef[n]

    for k in range(1, n + 1):
        p = coef[n - k] + (x - x_data[n - k]) * p

    return p
```

```python
x = np.array([-5, -1, 0, 2])
y = np.array([-2, 6, 1, 3])

# Menghitung koefisien divided difference
a_s = divided_diff(x, y)[0, :]

# Evaluasi interpolasi pada titik-titik baru
x_new = np.arange(-5, 2.1, .1)
y_new = newton_poly(a_s, x, x_new)

# Membuat plot
plt.figure(figsize=(12, 8))
plt.plot(x, y, 'bo')  # Menampilkan titik data asli
plt.plot(x_new, y_new)  # Menampilkan kurva interpolasi
```

```python
import numpy as np

def divided_diff(x, y):
    n = len(y)
    coef = np.zeros([n, n])
    coef[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            coef[i][j] = (coef[i + 1][j - 1] - coef[i][j - 1]) / (x[i + j] - x[i])

    return coef

def newton_poly(coef, x_data, x):
    n = len(x_data) - 1
    p = coef[n]

    for k in range(1, n + 1):
        p = coef[n - k] + (x - x_data[n - k]) * p

    return p


x_data = np.array([0, 5])
y_data = np.array([10, 15])

coef = divided_diff(x_data, y_data)[0, :]

x_interp = 7.2

# Menghitung kecepatan pada jarak 7.2 meter
v_7_2 = newton_poly(coef, x_data, x_interp)

print(f"Kecepatan Ben Johnson pada jarak {x_interp} meter adalah {v_7_2:.2f} m/detik")

x_range = np.linspace(0, 10, 100)
y_interp = newton_poly(coef, x_data, x_range)

plt.figure(figsize=(8, 6))
plt.plot(x_data, y_data, 'bo', label='Data Points')
plt.plot(x_range, y_interp, 'r-', label='Interpolated Curve')


plt.title('Ben Johnson\'s Speed Interpolation')
plt.xlabel('Distance (meters)')
plt.ylabel('Speed (m/s)')
plt.show()
```
